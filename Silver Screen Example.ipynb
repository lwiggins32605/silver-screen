```python
value_counts = data_df['Genres'].value_counts(dropna=True, sort=True)

df = value_counts.rename_axis('Genres').reset_index(name='counts')

# df.head()

plt.bar(df["Genres"], df["counts"])
```

```python
production_data = data_df["Production Company"]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]
explode = (0.1, 0, 0, 0, 0)  
plt.pie(production_data, labels=production_data, explode=explode, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Gold medal achievements of five most successful\n"+"countries in 2016 Summer Olympics")
plt.show()
```
